# `geom_gauge()`

`geom_gauge()` draws a semicircle gauge at a given `x`/`y` position.

Use it when you need compact KPI-like indicators in dashboards.
At the moment, gauge `value` and `width` are layer parameters (not mapped aesthetics), so they should be set explicitly for each layer.

```python
import pandas as pd
from lets_plot import *
```

```python
LetsPlot.setup_html(no_js=True)
```

## 1. Basic gauge

```python
basic_data = {'x': [0], 'y': [0], 'label': ['72%']}

(
    ggplot(basic_data, aes('x', 'y')) +
    geom_gauge(value=0.72, size=46, width=12, fill='#2E7D32') +
    geom_text(aes(label='label'), nudge_y=0.15, size=16, fontface='bold') +
    coord_fixed() +
    xlim(-1, 1) + ylim(-0.2, 1.0) +
    flavor_darcula()#theme_void() 
)
```

## 2. `value` controls the filled arc

```python
def gauge_by_value(value, title):
    return (
        ggplot({'x': [0], 'y': [0]}, aes('x', 'y')) +
        geom_gauge(value=value, size=20, width=10, fill='#1565C0') +
        geom_text(
            aes('x', 'y', label='label'),
            data={'x': [0], 'y': [0.14], 'label': [f"{int(value * 100)}%"]},
            size=13
        ) +
        coord_fixed() +
        xlim(-1, 1) + ylim(-0.2, 1.0) +
        ggtitle(title) +
        theme_void() +
        theme(plot_title=element_text(hjust=0.5))
    )

gggrid([
    gauge_by_value(0.2, 'Low'),
    gauge_by_value(0.55, 'Medium'),
    gauge_by_value(0.9, 'High'),
], ncol=3)
```

## 3. Gauge thickness (`width`) and radius (`size`)

```python
def gauge_shape(width, size, title):
    return (
        ggplot({'x': [0], 'y': [0]}, aes('x', 'y')) +
        geom_gauge(value=0.65, width=width, size=size, fill='#F57C00') +
        coord_fixed() +
        xlim(-1, 1) + ylim(-0.2, 1.0) +
        ggtitle(title) +
        theme_void() +
        theme(plot_title=element_text(hjust=0.5))
    )

gggrid([
    gauge_shape(width=4, size=20, title='width=4, size=20'),
    gauge_shape(width=20, size=20, title='width=20, size=20'),
    gauge_shape(width=4, size=30, title='width=4, size=30'),
    gauge_shape(width=20, size=30, title='width=20, size=30'),
], ncol=2)
```

## 4. Small KPI board with multiple gauges

```python
kpi_df = pd.DataFrame({
    'metric': ['Availability', 'Quality', 'Delivery'],
    'x': [0.0, 1.5, 3.0],
    'y': [0.0, 0.0, 0.0],
    'value': [0.94, 0.78, 0.61],
    'fill': ['#2E7D32', '#00838F', '#EF6C00'],
    'radius': [15, 15, 15],
    'width': [8, 8, 8],
})
kpi_df
```

```python
p = (
    ggplot() +
    coord_fixed() +
    xlim(-0.8, 3.8) + ylim(-0.35, 1.05) +
    theme_void()
)

for row in kpi_df.itertuples(index=False):
    p = p + geom_gauge(
        aes('x', 'y'),
        data={'x': [row.x], 'y': [row.y]},
        value=row.value,
        size=row.radius,
        width=row.width,
        fill=row.fill
    )
    p = p + geom_text(
        aes('x', 'y', label='label'),
        data={'x': [row.x], 'y': [0.18], 'label': [f"{int(row.value * 100)}%"]},
        size=11,
        fontface='bold'
    )
    p = p + geom_text(
        aes('x', 'y', label='label'),
        data={'x': [row.x], 'y': [-0.22], 'label': [row.metric]},
        size=9
    )

p + ggtitle('Team KPI Gauges') + theme(plot_title=element_text(hjust=0.5))
```

## Notes

- `value` should be in `[0, 1]`.
- `width` is measured in pixels.
- `size` controls gauge radius.
- To show different values in one chart, add separate `geom_gauge()` layers (as in the KPI example).
